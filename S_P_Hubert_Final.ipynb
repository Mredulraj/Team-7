```python
import torch
from transformers import Wav2Vec2Processor, HubertForCTC
import soundfile as sf

processor = Wav2Vec2Processor.from_pretrained("facebook/hubert-xlarge-ls960-ft")
model = HubertForCTC.from_pretrained("facebook/hubert-xlarge-ls960-ft")

# processor = HubertForCTC.from_pretrained("facebook/hubert-base-ls960")

# model = HubertForCTC.from_pretrained("facebook/hubert-base-ls960")

def map_to_array(batch):
    speech, _ = sf.read(batch["file"])
    batch["speech"] = speech
    return batch
```

```python
import torch

train_=[]

def process(text,lim):
  start=0
  while lim>0:
    print(lim)
    yield processor(text[start],return_tensors="pt").input_values
    
    lim-=1
    start+=1

process(train,2)
```

```python
train_=[]
for _ in process(train,np.shape(train)[0]):
  train_.append(_)
```

```python
len(train_)
```

```python
torch.cuda.empty_cache()

device = "cuda:0" if torch.cuda.is_available else "cpu"

model= model.to(device)


def model_(text,lim1,lim2):
  lim1=int(lim1)
  lim2=int(lim2)
  
  while lim2>0:
    print(lim2)
    with torch.no_grad():

      a=text[lim1].to(device)
      b=model(a)

      yield processor.decode(torch.argmax(b.logits,dim=-1)[0])

      #yield processor.decode(torch.argmax(model(text[start].to(device)).logits, dim=-1)[0])
    del a,b    
    torch.cuda.empty_cache()

    lim2-=1
    lim1+=1

model_(train_,10,20)
```
